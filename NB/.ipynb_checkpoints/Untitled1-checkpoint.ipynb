```python
%config Completer.use_jedi = False  # To avoid delays in autocompletion during coding.
import os
import sys
os.chdir('..')  # changing to a higher dir to access custom modules
cwd = os.getcwd()
sys.path.append(cwd)

import numpy as np  # numpy for matrix, linear-algebra, and other numerical studies.
import warnings  # to send warnings to user.
import pandas as pd  # pandas for handling data similar to spreadsheet.
import matplotlib as mpl  # matplotlib for plots
import matplotlib.pyplot as plt  # for plots
import seaborn as sb  # for styling of plots
from PyImModules.matplotlib_settings import *
from itertools import product
from scipy.ndimage import gaussian_filter1d
from scipy.interpolate import interp1d
mpl.rcParams['figure.dpi'] = 150  # Reducing resolution of plots for better viewability in notebook.
```

```python
# 3D isotropic orientation state of fibres
Nfibres = 1000
phiDistr = np.random.uniform(low=0, high=np.pi, size=Nfibres)
u  = np.random.rand(Nfibres)
thetaDistr = np.arccos(1 - 2*u)

alphaDistr = np.pi/2 + np.arctan(np.tan(thetaDistr) * np.cos(phiDistr))

print("\nSummary of phi: ")
print("min: {}, max: {}, mean: {}".format(phiDistr.min(), phiDistr.max(), phiDistr.mean()))
print("\nSummary of theta: ")
print("min: {}, max: {}, mean: {}".format(thetaDistr.min(), thetaDistr.max(), thetaDistr.mean()))
print("\nSummary of alpha: ")
print("min: {}, max: {}, mean: {}".format(alphaDistr.min(), alphaDistr.max(), alphaDistr.mean()))
```

```python
fig, axes = plt.subplots(3, 1, figsize=(3, 7))
axs = axes.ravel()
bins = np.deg2rad(np.arange(0, 181, 10))

# Phi
ax = axs[0]
h, b, _ = ax.hist(phiDistr, bins=bins, density=True)
ax.set_xlabel('$\phi$')
ax.set_ylabel('$p(\phi)$')
ax.set_xticks(b[::3])

# Theta
ax = axs[1]
h, b, _ = ax.hist(thetaDistr, bins=bins, density=True)
ax.set_xlabel('$\\theta$')
ax.set_ylabel('$p(\\theta)$')
ax.set_xticks(b[::3])

# Theta
ax = axs[2]
h, b, _ = ax.hist(alphaDistr, bins=bins, density=True)
ax.set_xlabel('$\\alpha$')
ax.set_ylabel('$p(\\alpha)$')
ax.set_xticks(b[::3])

for ax in axs:
    ax.set_ylim([0, 0.55])

fig.suptitle('Orientation distribution')
plt.show()
```

```python
# Theoretical ODFs
phiVals = np.deg2rad(np.arange(0.5, 180))
dphi = np.mean(phiVals[1:] - phiVals[:-1])
phiPMF  = dphi * np.array([1 / np.pi] * len(phiVals))

thetaVals = phiVals
dtheta = np.mean(thetaVals[1:] - thetaVals[:-1])
thetaPMF  = (1 - np.cos(dtheta)) * np.cos(thetaVals) + np.sin(dtheta) * np.sin(thetaVals)
thetaPMF  = 0.5 * thetaPMF

fig = plt.figure(figsize=(3.5, 2))
ax = fig.gca()
ax.plot(phiVals, phiPMF, lw=0.75, label="$p(\phi)$")
ax.plot(thetaVals, thetaPMF, lw=0.75, label="$p(\\theta)$")
ax.legend()
```

```python
def map_rv2cos(rv_in, pmf_in, rv_out):
    assert len(rv_in) == len(pmf_in), print("check input: len(rv_in) not equal to len(pmf_in).")
    
    cdf_in = np.cumsum(pmf_in)
    F = interp1d(rv_in, cdf_in, kind='quadratic')
    
    cdf_out = F(np.arccos(-rv_out))
    pmf_out = np.gradient(cdf_out)
    assert len(pmf_out) == len(rv_out)
    
    return pmf_out


def map_rv2tan(rv_in, pmf_in, rv_out):
    assert len(rv_in) == len(pmf_in), print("check input: len(rv_in) not equal to len(pmf_in).")
    
    cdf_in = np.cumsum(pmf_in)
    F = interp1d(rv_in, cdf_in, kind='quadratic')
    
    cdf_out = F(np.arctan(rv_out))
    pmf_out = np.gradient(cdf_out)
    assert len(pmf_out) == len(rv_out)
    
    return pmf_out
```

```python
alpha_mat = np.array([np.arctan(np.tan(th) * np.cos(ph)) for ph, th in product(phiVals, thetaVals)])
alpha_mat = alpha_mat.reshape((len(phiVals), len(thetaVals)))
print("min(alpha): {0},\t max(alpha): {1}".format(alpha_mat.min(), alpha_mat.max()))

p_joint = np.array([p1 * p2 for p1, p2 in product(phiPMF, thetaPMF)])
p_joint = p_joint.reshape((len(phiPMF), len(thetaPMF)))

alphaVals = np.deg2rad(np.arange(-90, 90))
dalpha = np.mean(alphaVals[1:] - alphaVals[:-1])

alphaPMF = []
for a in alphaVals:
    p = 0
    for phidx, ph in enumerate(phiVals):
        row = alpha_mat[phidx, :]
        mask = (a < row) & (row <= a + dalpha)
        p += np.sum(p_joint[phidx, :] * mask) / np.abs(ph)
    alphaPMF.append(p)
print("Checking total proability: ", np.sum(alphaPMF))
alphaPMF = gaussian_filter1d(alphaPMF, 2)
print("Checking total proability: ", np.sum(alphaPMF))

fig = plt.figure()
ax = fig.gca()
ax.plot(alphaVals, alphaPMF, lw=0.75, label='$p(\\alpha)$')
```
